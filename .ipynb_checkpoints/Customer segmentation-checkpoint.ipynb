```python
#FILL THE MISSING VALUES WITH THE MEDIAN VALUES.. 
df['Income']=df['Income'].fillna(df['Income'].median())
```

```python
df.isna().sum().sort_values(ascending =False)
```

```python
df['Dt_Customer'] = pd.to_datetime(df.Dt_Customer)
```

```python
df=df.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish",
                          "MntSweetProducts":"Sweets","MntGoldProds":"Gold","Dt_Customer":"Register",
                          "NumDealsPurchases":"Deals","NumWebPurchases":"Web","NumCatalogPurchases":"Catalog",
                         "NumStorePurchases":"Store","NumWebVisitsMonth":"WebVisits","AcceptedCmp3":"Cmp3",
                         "AcceptedCmp4":"Cmp4","AcceptedCmp1":"Cmp1","AcceptedCmp2":"Cmp2","AcceptedCmp5":"Cmp5"})

#REPLACING THE CONFLICT VALUES IN Marital_status..
df['Marital_Status'] = df['Marital_Status'].replace(['Married', 'Together'],'In_Relationship')
df['Marital_Status'] = df['Marital_Status'].replace(['Divorced', 'Widow', 'Alone', 'YOLO', 'Absurd'],'Single')

#CHANGING CATEGORY INTO "UG" AND "PG" ONLY....
df['Education'] = df['Education'].replace(['PhD','2n Cycle','Graduation', 'Master'],'Post Graduate')  
df['Education'] = df['Education'].replace(['Basic'], 'Under Graduate')

#create age of customer
df["Age"] = 2014-df["Year_Birth"]

#Total spendings on various items
df["Total_Spent"] = df["Wines"]+ df["Fruits"]+ df["Meat"]+ df["Fish"]+ df["Sweets"]+ df["Gold"]

#Total child
df['Children']=df['Kidhome']+df['Teenhome']

#Feature pertaining parenthood
df["Is_Parent"] = np.where(df['Children']> 0, 1, 0)

#Total accepted cmps
df['Accepted']=df['Cmp1']+df['Cmp2']+df['Cmp3']+df['Cmp4']+df['Cmp5']
```

```python
#Creating a copy of data
ds = df.copy()
#Dropping some of the redundant features
to_drop = ["Marital_Status", "Register", "Year_Birth", "ID"]
ds = ds.drop(to_drop, axis=1)
```

```python
# creating a subset of dataframe by dropping the features on deals accepted and promotions
cols_del = ['Cmp3', 'Cmp4', 'Cmp5', 'Cmp1','Cmp2', 'Complain', 'Response']
ds = ds.drop(cols_del, axis=1)
#Scaling
scaler = StandardScaler()
scaler.fit(ds)
scaled_ds = pd.DataFrame(scaler.transform(ds),columns= ds.columns )
print("All features are now scaled")
```

```python
#Scaled data to be used for reducing the dimensionality
print("Dataframe to be used for further modelling:")
scaled_ds.head(10)
```

# DIMENSIONALITY REDUCTION

```python
#Initiating PCA to reduce dimentions aka features to 3
pca = PCA(n_components=3)
pca.fit(scaled_ds)
PCA_ds = pd.DataFrame(pca.transform(scaled_ds), columns=(["col1","col2", "col3"]))
PCA_ds.describe().T
```

```python
#A 3D Projection Of Data In The Reduced Dimension
x =PCA_ds["col1"]
y =PCA_ds["col2"]
z =PCA_ds["col3"]
#To plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z, c="maroon", marker="o" )
ax.set_title("A 3D Projection Of Data In The Reduced Dimension")
plt.show()
```

# Elbow

```python
# elbow method

inertia= []
k= np.arange(1,9)
for i in k:
    model = KMeans(n_clusters= i, max_iter=5)
    ab = model.fit(PCA_ds)
    inertia.append(pd.Series({'k': i,
                              'inertia': model.inertia_
                             }))
inertias= pd.concat(inertia, axis=1).T.set_index('k')
# draw elbow
plt.figure(figsize=(12,6))
plt.plot(inertias.index, inertias['inertia'], '-*')
plt.title('Best k using Elbow method')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(k)
plt.show()
```

```python
# Quick examination of elbow method to find numbers of clusters to make.
print('Elbow Method to determine the number of clusters to be formed:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(PCA_ds)
Elbow_M.show()
```

# K-MEANS CLUSTERING

```python
model = KMeans(n_clusters=4, random_state=0).fit(PCA_ds)
pr_AC = model.predict(PCA_ds)
PCA_ds["Clusters"] = pr_AC
#Adding the Clusters feature to the orignal dataframe.
df["Clusters"]= pr_AC
```

```python
#Plotting the clusters
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(x, y, z, s=40, c=PCA_ds["Clusters"], marker='o', cmap = cmap )
ax.set_title("The Plot Of The Clusters")
plt.show()
```

# Silhouette

```python
score=silhouette_score(PCA_ds, PCA_ds["Clusters"], metric = "euclidean")
print("KMeans Silhouette score:", score)
```
